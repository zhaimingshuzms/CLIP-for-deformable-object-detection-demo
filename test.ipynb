```python
import torch
import torch.nn.functional as F
alpha = torch.ones(4)
alpha[0]=3
alpha.requires_grad_()
a = F.normalize(alpha,dim=0).log().float()
KLdivergence = torch.nn.KLDivLoss(reduction='sum')
loss = KLdivergence(a,torch.ones(4))
loss.backward()
```
